```python
# The geometry column gets lost in the spatial join
gdp['g1'] = gdp.geometry
states['g2'] = states.geometry
districts['g2'] = districts.geometry
```

```python
intersection = gpd.sjoin(gdp, districts, 
                         how="inner", 
                         predicate='intersects')
```

```python
intersection['intersect_area'] = intersection.g1.intersection(intersection.g2).to_crs('EPSG:7755').area
```

```python
intersection['grid_area'] = intersection.g1.to_crs('EPSG:7755').area
```

```python
intersection['pct_intersect'] = intersection['intersect_area']/intersection['grid_area']
```

Coastal Grids like `7986` show <1 intersect with polygons. But their entire GDP only belongs to the mainland (Cannot give GDP to sea)
To consider such grids, we normalise the pct_intersect so that sum of it equals 1.

```python
# Normalize pct by group
intersection['pct_intersect_normalised'] = intersection['pct_intersect'] / intersection.groupby('fid')['pct_intersect'].transform('sum')
```

```python
intersection[intersection.fid == 7986]
```

```python
# GDP by State
# First compute the weighted value
intersection['weighted_GDP'] = intersection['DN'] * intersection['pct_intersect_normalised'] / 1e9 #Billion PPP Dollars
```

```python
gdp_by_state = intersection.groupby('stname')['weighted_GDP'].sum().reset_index()
gdp_by_state.to_csv('GDP_by_State.csv',index=False)
```

```python
gdp_by_district = intersection.groupby('stdtname')['weighted_GDP'].sum().reset_index()
gdp_by_district = districts.merge(gdp_by_district, on='stdtname')#.to_csv('GDP_by_District.csv',index=False)
gdp_by_district.drop(['geometry', 'g2'], axis=1).to_csv('GDP_by_District.csv',index=False)
```
